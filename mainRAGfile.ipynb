```python
# Sanity check: verify evaluation output
import json
with open("eval_results/marker_report_evaluation.json", encoding="utf-8") as f:
    results = json.load(f)
print(f"Total entries: {len(results)}")
for r in results:
    rid = r["id"]
    has_report = "report" in r
    has_baseline = "baseline" in r
    report_err = "error" in r.get("report", {}).get("evaluation", {})
    baseline_err = "error" in r.get("baseline", {}).get("evaluation", {}) if has_baseline else "N/A"
    score = r["report"]["evaluation"].get("overall_score", "ERR") if not report_err else "ERR"
    print(f"  {rid}: report={has_report} baseline={has_baseline} score={score} report_err={report_err} baseline_err={baseline_err}")
```
